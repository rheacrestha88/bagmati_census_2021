```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/2021_Nepal_census'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
soup
```

```python
## finding the table which you want to extract from the web
## Population of Bagmati Province 

census_table = soup.find_all('table')[6]
```

```python
census_col = census_table.find_all('th')
```

```python
##looping through the texts present between th

bagmati_census_table = [title.text.strip('\n') for title in census_col]
bagmati_census_table
```

```python
## creating column dataframe

df= pd.DataFrame(columns= bagmati_census_table)
df
```

```python
column_data = census_table.find_all('tr')
```

```python
for row in column_data[1: ]:
    row_data = row.find_all('td')
    individual_data = [data.text.strip('\n') for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_data
```

```python
df.to_csv('Bagmati_population_census_2021.csv', index= False)
```
